## Helper Functions

Below are some helper functions to help get you started. They should look familiar from the lesson!

````python
import math

def grayscale(img):
    """Applies the Grayscale transform
    This will return an image with only one color channel
    but NOTE: to see the returned image as grayscale
    (assuming your grayscaled image is called 'gray')
    you should call plt.imshow(gray, cmap='gray')"""
    return cv2.cvtColor(img, cv2.COLOR_RGB2GRAY)
    # Or use BGR2GRAY if you read an image with cv2.imread()
    # return cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
    
def canny(img, low_threshold, high_threshold):
    """Applies the Canny transform"""
    return cv2.Canny(img, low_threshold, high_threshold)

def gaussian_blur(img, kernel_size):
    """Applies a Gaussian Noise kernel"""
    return cv2.GaussianBlur(img, (kernel_size, kernel_size), 0)

def region_of_interest(img):
    """
    Applies an image mask.
    
    Only keeps the region of the image defined by the polygon
    formed from `vertices`. The rest of the image is set to black.
    """
    #defining a blank mask to start with
    mask = np.zeros_like(img)   
    
    imshape = img.shape #extracts y and x size
    
    #defining region of interest ratios. It is necessary because of the difference in size in test videos and challenge video
    ax = 115 / 960
    ay = 1
    
    bx = 435 / 960
    by = 330 / 540 
    
    cx = 550 / 960 
    cy = 330 / 540 
    
    dx = 900 / 960 
    dy = 1
    
    vertices = np.array([[(ax*imshape[1], ay*imshape[0]), (bx*imshape[1], by*imshape[0]), (cx*imshape[1], cy*imshape[0]), (dx*imshape[1], dy*imshape[0])]], dtype=np.int32)
    
    #defining a 3 channel or 1 channel color to fill the mask with depending on the input image
    if len(img.shape) > 2: # testing for number of channels
        channel_count = img.shape[2]  # i.e. 3 or 4 depending on your image
        ignore_mask_color = (255,) * channel_count # red colour
    else:
        ignore_mask_color = 255 # red colour
        
    #filling pixels inside the polygon defined by "vertices" with the fill color    
    cv2.fillPoly(mask, vertices, ignore_mask_color)
    
    #returning the image only where mask pixels are nonzero
    masked_image = cv2.bitwise_and(img, mask)
    return masked_image

def draw_lines(img, lines, color=[255, 0, 0], thickness=8): 
    """
    Function draws Hough lines.
    """
    
    for line in lines:     
        for x1,y1,x2,y2 in line:
            cv2.line(img, (x1, y1), (x2, y2), color, thickness)


def draw_lines_cont(img, lines, color=[255, 0, 0], thickness=8):
    """
    Function draws a one continuos line out od Hough lines.
    """
    # defining initial conditions for line vertices optimization
    x1l = img.shape[1]
    x2l = 0
    y1l = 0
    y2l = img.shape[0]
    
    x1r = img.shape[1]
    x2r = 0
    y1r = img.shape[0]
    y2r = 0
    
    for line in lines:
       
        for x1,y1,x2,y2 in line:
            # sorting the lines to left and right line
            if ((y2-y1)/(x2-x1)) < -0.55: #and ((y2-y1)/(x2-x1)) > -0.85: #left line
                if x1l > x1:
                    x1l = x1
                if y1l < y1:
                    y1l = y1
                if x2l < x2:
                    x2l = x2
                if y2l > y2:
                    y2l = y2

            elif ((y2-y1)/(x2-x1)) > 0.55: #and ((y2-y1)/(x2-x1)) < 0.85: #right line
                if x1r > x1:
                    x1r = x1
                if y1r > y1:
                    y1r = y1
                if x2r < x2:
                    x2r = x2
                if y2r < y2:
                    y2r = y2
   
    # defining polinomial coeficients 
    B1 = (y2l*x1l - y1l*x2l) / (x1l - x2l)
    A1 = (y1l - B1) / x1l  
    B2 = (y2r*x1r - y1r*x2r) / (x1r - x2r)
    A2 = (y1r - B2) / x1r  
    
    # defining ratios
    by = 330/540 #330
    cy = 330/540 #330
    
    cv2.line(img, (int((img.shape[0] - B1) / A1), img.shape[0]), (int((by*img.shape[0] - B1) / A1), int(by*img.shape[0])), color, thickness)
    cv2.line(img, (int((cy*img.shape[0] - B2)/A2), int(cy*img.shape[0])), (int((img.shape[0] - B2)/A2), img.shape[0]), color, thickness)
            
def draw_lines_cont2(img, lines, color=[255, 0, 0], thickness=8):
    """
    Function draws a one continuos line out od Hough lines.
    """
    # Attempt to improve the pipeline for the challenge video
    
    
    # defining initial conditions for line vertices optimization
    a1 = 0
    b1 = 0
    i1 = 0
    a2 = 0
    b2 = 0
    i2 = 0
    
    for line in lines:
       
        for x1,y1,x2,y2 in line:
            # sorting the lines to left and right line
            if ((y2-y1)/(x2-x1)) < -0: #left line
                b1 += (y2*x1 - y1*x2) / (x1 - x2)
                a1 += (y1 - (y2*x1 - y1*x2) / (x1 - x2)) / x1 
                i1 += 1
                
            elif ((y2-y1)/(x2-x1)) > 0: #right line
                b2 += (y2*x1 - y1*x2) / (x1 - x2)
                a2 += (y1 - (y2*x1 - y1*x2) / (x1 - x2)) / x1 
                i2 += 1
   
    # calculating polinomial coefficients 
    B1 = b1 / i1
    A1 = a1 / i1  
    B2 = b2 / i2
    A2 = a2 / i2  
    
    # defining ratios
    by = 330/540 #330
    cy = 330/540 #330
    
    cv2.line(img, (int((img.shape[0] - B1) / A1), img.shape[0]), (int((by*img.shape[0] - B1) / A1), int(by*img.shape[0])), color, thickness)
    cv2.line(img, (int((cy*img.shape[0] - B2)/A2), int(cy*img.shape[0])), (int((img.shape[0] - B2)/A2), img.shape[0]), color, thickness)
            

def hough_lines(img, rho, theta, threshold, min_line_len, max_line_gap):
    """
    `img` should be the output of a Canny transform.
        
    Returns an image with hough lines drawn.
    """
    lines = cv2.HoughLinesP(img, rho, theta, threshold, np.array([]), minLineLength=min_line_len, maxLineGap=max_line_gap)
    line_img = np.zeros((img.shape[0], img.shape[1], 3), dtype=np.uint8)
    draw_lines(line_img, lines)
    return line_img

def hough_lines_cont(img, rho, theta, threshold, min_line_len, max_line_gap):
    """
    `img` should be the output of a Canny transform.
        
    Returns an image with hough lines drawn.
    """
    lines = cv2.HoughLinesP(img, rho, theta, threshold, np.array([]), minLineLength=min_line_len, maxLineGap=max_line_gap)
    line_img = np.zeros((img.shape[0], img.shape[1], 3), dtype=np.uint8)
    draw_lines_cont(line_img, lines)
    return line_img

# Python 3 has support for cool math symbols.

def weighted_img(img, initial_img, =0.8, =1., =0.):
    """
    `img` is the output of the hough_lines(), An image with lines drawn on it.
    Should be a blank image (all black) with lines drawn on it.
    
    `initial_img` should be the image before any processing.
    
    The result image is computed as follows:
    
    initial_img *  + img *  + 
    NOTE: initial_img and img must be the same shape!
    """
    return cv2.addWeighted(initial_img, , img, , )
````

````python
def process_image(image):
    """
    Pipeline for processing videos
    """

    gray = grayscale(image)
    gray_blurred = gaussian_blur(gray, 5) #kernel size is 5. Gaussian Blur i LPF that removes uneccesary edges in the image
    edge = canny(gray_blurred, 40, 120) #Recommended values for low and high Canny treshold are 1/3. Possible values for an 8bit 
    #grayscale image are 0:255 (0 - black, 255 - white)
    
    masked_canny = region_of_interest(edge) #on grayscale mask will only let through white colour
    
    #rho = 1 # distance resolution in pixels of the Hough grid
    theta = np.pi/180 # angular resolution in radians of the Hough grid
    threshold = 40 # minimum number of intersections in Hough grid for a line to be declared in polar grid
    min_line_len = 50 # minimum number of pixels making up a line
    max_line_gap = 100 # maximumm gap between pixels between connectable line elements

    lines = hough_lines_cont(masked_canny, rho, theta, threshold, min_line_len, max_line_gap) # output lines is an array 
    # containing endpoints of detected line elements

    lines_weighted = weighted_img(lines, image)
    
    return lines_weighted
````

````python
def process_image_robust(image):
    """
    Robust pipeline for processing videos
    """
   
    gray = grayscale(image)
    gray_blurred = gaussian_blur(gray, 5) #kernel size is 5. Gaussian Blur i LPF that removes uneccesary edges in the image
    edge = canny(gray_blurred, 40, 60) #Recommended values for low and high Canny treshold are 1/3. Possible values for an 8bit
    # grayscale image are 0:255 (0 - black, 255 - white)
    
    masked_canny = region_of_interest(edge) #on grayscale mask will only let through white colour
    
    rho = 1 # distance resolution in pixels of the Hough grid
    theta = np.pi/180 # angular resolution in radians of the Hough grid
    threshold = 30 # minimum number of intersections in Hough grid for a line to be declared in polar grid
    min_line_len = 40 # minimum number of pixels making up a line
    max_line_gap = 100 # maximumm gap between pixels between connectable line elements

    lines = hough_lines_cont(masked_canny, rho, theta, threshold, min_line_len, max_line_gap) # output lines is an array 
    # containing  endpoints of detected line elements

    lines_weighted = weighted_img(lines, image)
    
    return lines_weighted
````
